```python
import numpy as np

data = np.array([12, 15, 18, 19, 20, 22, 25, 100])

q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1

lower_cut = q1 - 1.5 * iqr
upper_cut = q3 + 1.5 * iqr

outliers = data[(data < lower_cut) | (data > upper_cut)]
print("Outliers:", outliers)
```

```python
filtered_data = data[(data >= lower_cut) & (data <= upper_cut)]

new_mean = filtered_data.mean()
new_std = filtered_data.std()

print("Clean Mean:", new_mean)
print("Clean Std Dev:", new_std)
```

```python
def descriptive_stats(data):
    n = len(data)

    # Mean
    total = 0
    for v in data:
        total += v
    mean = total / n

    # Median
    sorted_data = sorted(data)
    mid = n // 2
    if n % 2 == 0:
        median = (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        median = sorted_data[mid]

    # Mode
    freq = {}
    for v in data:
        freq[v] = freq.get(v, 0) + 1
    mode = max(freq, key=freq.get)

    # Standard Deviation
    variance_sum = 0
    for v in data:
        variance_sum += (v - mean) ** 2
    std_dev = (variance_sum / n) ** 0.5

    return mean, median, mode, std_dev


dataset = [2, 4, 6, 6, 8, 10, 50]
print(descriptive_stats(dataset))
```

```python
import numpy as np

x = np.array([1, 2, 3, 4, 100])
y = np.array([2, 4, 6, 8, 200])

def iqr_clean(arr):
    q1 = np.percentile(arr, 25)
    q3 = np.percentile(arr, 75)
    spread = q3 - q1
    return arr[(arr >= q1 - 1.5 * spread) & (arr <= q3 + 1.5 * spread)]

x_clean = iqr_clean(x)
y_clean = iqr_clean(y)

corr_before = np.corrcoef(x, y)[0, 1]
corr_after = np.corrcoef(x_clean, y_clean)[0, 1]

print("Correlation before:", corr_before)
print("Correlation after:", corr_after)
```

```python
# Import Libraries
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing # Changed from load_boston
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score
from scipy import stats

# Load Dataset
housing = fetch_california_housing() # Changed from boston = load_boston()
X = pd.DataFrame(housing.data, columns=housing.feature_names)
y = pd.Series(housing.target, name=housing.target_names[0]) # Adjusted target name for California Housing

# Outlier removal using Z-score
z_scores = np.abs(stats.zscore(X))
outlier_rows = (z_scores > 3).any(axis=1)
X_clean = X[~outlier_rows]
y_clean = y[~outlier_rows]

print(f"Original dataset size: {X.shape[0]}, After outlier removal: {X_clean.shape[0]}")

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42)

# Build Pipeline: Scaling + Regression
pipeline = Pipeline([
    ('scaler', StandardScaler()),      # Step 1: Scale features
    ('regressor', LinearRegression())  # Step 2: Linear Regression
])

# Train the pipeline
pipeline.fit(X_train, y_train)

# Predict on test set
y_pred = pipeline.predict(X_test)

# Evaluate Performance
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("\nRegression Performance Metrics:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R2 score): {r2:.2f}")
```
